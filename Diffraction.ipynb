```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from scipy.special import jve
import matplotlib as mpl
import matplotlib.animation as animation
from matplotlib.colors import ListedColormap
```

```python
D = 1 # (distance of the screen) m
wl =10**(-7) # m
d = 10**(-6) # m (width of slit)

def get_intensity(x, D, wl, d):
    theta_0 = np.arctan(x/D)
    X = (d* np.pi * np.sin(theta_0))/(wl)
    intensity =(np.sin(X)/X)**2
    return(intensity)
rlist = np.linspace(10**(-8),0.5,1000)
ir1 = get_intensity(rlist, D, wl, d)
plt.plot(rlist,ir1)

thetalist=np.radians(np.arange(0,361,1)) #Radians
rmesh, thetamesh = np.meshgrid(rlist, thetalist)

FullFunction = get_intensity(rmesh, D, wl, d)
ln_ir1 = np.log(ir1)
FullFunction2 = FullFunction**2*(ln_ir1)**2

fig, ax = plt.subplots(dpi=120,subplot_kw=dict(projection='polar'))
plt.axis('off')
gray_map = mpl.colormaps.get_cmap('gray')
newcmp = ListedColormap(gray_map(np.linspace(0.25, 1, 256)))

ax.contourf(thetamesh, rmesh, FullFunction, 100, cmap=newcmp)
```
